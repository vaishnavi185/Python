```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([0, 1, 9, 3, 4, 5],kind="kde")  #create distribution plot take input as array  ,kind can be "kde" or "hist"

plt.show()
```

```python
#normal distribution plot it take mean , loc and scale as input
data = np.random.normal(size=100)
sns.displot(data,kind="kde")
plt.show()
```

```python
#binomila distribution --> for discrete data
data =np.random.binomial(n=1,p=0.5,size=100)
sns.displot(data,kind="hist")
plt.show()
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

data = {
  "normal": random.normal(loc=50, scale=5, size=1000),
  "binomial": random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

```python
data = {
    "days":["mon","tue","wed","thru","fri"],
    "no of people":[12,9,14,12,10]
}
df= pd.DataFrame(data)
sns.lineplot(x="days",y="no of people",data=df)
plt.show()
```

```python
xx=sns.load_dataset("titanic")
df=pd.DataFrame(xx)
df.head()
df.tail()
df.info()
df.dropna(inplace=True)
sns.lineplot(x="age",y="fare",data=xx,hue="who",style="who")
plt.show()
```

```python
sns.barplot(x="who",y="age",data=xx,hue="sex",palette="spring",estimator=np.median ,order=["child","woman","man"])
plt.show()
```

```python
sns.histplot(data=xx,x="age",discrete=True,hue='who')

```

```python
sns.scatterplot(x="age",y="fare",data=xx,hue="who",alpha=0.5,marker="*")
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
sns.heatmap(data=data,cmap="mako",linecolor="white",linewidths=2,annot=True)
```

```python
sns.countplot(x="who",data=xx,hue="who")
```

```python
mp = sns.FacetGrid(xx, col="class") 
mp.map_dataframe(sns.histplot, x="who", hue="who")
mp.add_legend()
```

```python
x=np.random.randint(low=1,high=100,size =(10))
y=np.random.randint(low=1,high=100,size=(10))
sns.kdeplot(x,y,shade=True)
```
